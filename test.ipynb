```python
import matplotlib.pyplot as plt

initial_amount = 1_000_000  # $1,000,000
# initial_amount = 100_000  # $1,000,000
annual_increase = 0.10       # 10% per year
years = 20                   # Number of years to simulate

# Calculate yearly amounts
amounts = [initial_amount]
for year in range(1, years + 1):
    new_amount = amounts[-1] * (1 + annual_increase)
    amounts.append(new_amount)

# Generate list of years (0 to 'years')
years_list = list(range(years + 1))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(years_list, amounts, marker='o', linestyle='-', color='b')
plt.title("Growth of $1,000,000 at 10% Annual Increase")
plt.xlabel("Year")
plt.ylabel("Amount ($)")
plt.grid(True)
plt.xticks(years_list)  # Show every year on x-axis
plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation on y-axis

# Annotate the final amount
final_amount = amounts[-1]
plt.annotate(
    f"${final_amount:,.2f}",
    xy=(years, final_amount),
    xytext=(years - 5, final_amount * 0.7),
    arrowprops=dict(facecolor='black', shrink=0.05),
    fontsize=10,
)

plt.show()
```
